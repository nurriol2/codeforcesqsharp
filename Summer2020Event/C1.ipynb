```qsharp
 open Microsoft.Quantum.Measurement;
 operation Solve (qs : Qubit[]) : Unit {
        using (ancilla = Qubit()) {
            repeat {
                // Create |00‚ü© + |01‚ü© + |10‚ü© + |11‚ü© state
                ApplyToEach(H, qs);
                // Create (|00‚ü© + |01‚ü© + |10‚ü©) ‚äó |0‚ü© + |11‚ü© ‚äó |1‚ü©
                Controlled X(qs, ancilla);
                let res = MResetZ(ancilla);
            }
            until (res == Zero)
            fixup {
                ResetAll(qs);
            }
        }
    }
```
